```python
import pandas as pd

# =========================================
# LOAD DATASETS
# =========================================

train_df = pd.read_csv("../data/processed/train_labeled.csv")
val_df   = pd.read_csv("../data/processed/validation_labeled.csv")
test_df  = pd.read_csv("../data/processed/test_labeled.csv")

# =========================================
# VERIFY COLUMNS
# =========================================

print("=" * 60)
print("VERIFYING COLUMNS")
print("=" * 60)

expected_columns = [
    "meeting_id",
    "meeting_type",
    "split",
    "speaker",
    "text",
    "word_count",
    "action_item",
    "decision",
    "hedging_score",
    "topic"
]

print("\nTrain columns:")
print(list(train_df.columns))

missing_cols = [
    col for col in expected_columns
    if col not in train_df.columns
]

if len(missing_cols) == 0:
    print("\n✅ All expected columns are present.")
else:
    print("\n❌ Missing columns:")
    print(missing_cols)

# =========================================
# VERIFY SPLIT SIZES
# =========================================

print("\n" + "=" * 60)
print("VERIFYING SPLITS")
print("=" * 60)

print(f"\nTrain rows      : {len(train_df):,}")
print(f"Validation rows : {len(val_df):,}")
print(f"Test rows       : {len(test_df):,}")

# =========================================
# VERIFY CLASS DISTRIBUTION
# =========================================

print("\n" + "=" * 60)
print("CLASS DISTRIBUTION")
print("=" * 60)

print(f"\nTrain action-item rate      : {train_df['action_item'].mean() * 100:.2f}%")
print(f"Validation action-item rate : {val_df['action_item'].mean() * 100:.2f}%")
print(f"Test action-item rate       : {test_df['action_item'].mean() * 100:.2f}%")

# =========================================
# VERIFY NULL VALUES
# =========================================

print("\n" + "=" * 60)
print("NULL VALUE CHECK")
print("=" * 60)

print("\nNulls in train text :", train_df["text"].isna().sum())
print("Nulls in val text   :", val_df["text"].isna().sum())
print("Nulls in test text  :", test_df["text"].isna().sum())

# =========================================
# VERIFY MEETING OVERLAP
# =========================================

print("\n" + "=" * 60)
print("MEETING OVERLAP CHECK")
print("=" * 60)

train_meetings = set(train_df["meeting_id"].unique())
val_meetings   = set(val_df["meeting_id"].unique())
test_meetings  = set(test_df["meeting_id"].unique())

print("\nTrain-Val overlap :", len(train_meetings & val_meetings))
print("Train-Test overlap:", len(train_meetings & test_meetings))
print("Val-Test overlap  :", len(val_meetings & test_meetings))

# =========================================
# FINAL STATUS
# =========================================

print("\n" + "=" * 60)

if (
    len(missing_cols) == 0
    and train_df["text"].isna().sum() == 0
    and val_df["text"].isna().sum() == 0
    and test_df["text"].isna().sum() == 0
    and len(train_meetings & val_meetings) == 0
    and len(train_meetings & test_meetings) == 0
    and len(val_meetings & test_meetings) == 0
):
    print("✅ DATASET VERIFICATION PASSED")
else:
    print("❌ DATASET VERIFICATION FAILED")

print("=" * 60)
```
